```python
bunga['besar sepal'] = bunga['sepal length (cm)'] + bunga['sepal width (cm)']
bunga
#menambahkan kolom pada data 
#menambahkan 'besar sepal' sebagai kolom yang berisi data penjumlahan (sepal length (cm) + sepal width (cm) )
```

```python
bunga['besar sepal'] = bunga.iloc[:,4:7].sum(axis=1)
bunga
#'iloc' digunakan untuk mengetahui baris dari data 
#'sum(axis=1)' digunakan untuk operasi penjumlahan kolom
```

# Matplotlib

```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
```

# Grafik sederhana dengan Matplotlib

```python
bunga = pd.read_csv('https://raw.githubusercontent.com/imdwipayana/DATA/main/Bunga%20Iris/bunga_iris.csv')
print(bunga)

x = bunga['sepal length (cm)']
y = bunga['sepal width (cm)']
#menginisialisasikan variabel berdasarkan data yang diinginkan 

plt.plot(x,y)
#plot untuk menampilkan data koordinat pada garis koordinat 

plt.title('Grafik sepal')
#title untuk menamai grafik

plt.xlabel('X = panjang sepal')
plt.ylabel('Y = lebar setal')
#label untuk menamai sumbu yang ada pada grafik 
```

# Properti Pembuatan Grafik dengan Matplotlib

```python
x = bunga['sepal length (cm)']
y = bunga['petal length (cm)']
x1 = bunga['sepal width (cm)']
y1 = bunga['petal width (cm)']
#menginisialisasikan variabel dengan data yang dituju 


plt.figure(figsize=(10,10), dpi=200) 
#pembentukan sumbu koordinat dengan ukuran 10 x 10 inch dengan resolusi gambar 200 dpi
#'figure()' berguna untuk menggambarkan bentuk dari koordinatnya
#'figsize()' berguna untuk menentukan ukuran gambar

plt.plot(x,y, label = 'Grafik length' ,  color = 'purple' , 
         linewidth = 2 , marker = 'o' , markersize = 10 , 
         markeredgecolor = 'red' , linestyle='-')
#label untuk nama linier yang diminta
#linewidth untuk ketebalan linier
#marker untuk mejadi tanda pada titik yang dituju
#makersize unutuk ukuran tanda
#markeredgecolor untuk warna tanda

plt.plot(x1,y1, label = 'Garik width', marker = 'o' , 
         markersize = 10 , markeredgecolor = 'red')
x = bunga['sepal length (cm)']
y = bunga['petal length (cm)']
x1 = bunga['sepal width (cm)']
y1 = bunga['petal width (cm)']
#menginisialisasikan variabel dengan data yang dituju 


plt.figure(figsize=(10,10), dpi=200) 
#pembentukan sumbu koordinat dengan ukuran 10 x 10 inch dengan resolusi gambar 200 dpi
#'figure()' berguna untuk menggambarkan bentuk dari koordinatnya
#'figsize()' berguna untuk menentukan ukuran gambar

plt.plot(x,y, label = 'Grafik length' ,  color = 'purple' , 
         linewidth = 2 , marker = 'o' , markersize = 10 , 
         markeredgecolor = 'red' , linestyle='-')
#label untuk nama linier yang diminta
#linewidth untuk ketebalan linier
#marker untuk mejadi tanda pada titik yang dituju
#makersize unutuk ukuran tanda
#markeredgecolor untuk warna tanda

plt.plot(x1,y1, label = 'Garik width', marker = 'o' , 
         markersize = 10 , markeredgecolor = 'red')
x = bunga['sepal length (cm)']
y = bunga['petal length (cm)']
x1 = bunga['sepal width (cm)']
y1 = bunga['petal width (cm)']
#menginisialisasikan variabel dengan data yang dituju 


plt.figure(figsize=(10,10), dpi=200) 
#pembentukan sumbu koordinat dengan ukuran 10 x 10 inch dengan resolusi gambar 200 dpi
#'figure()' berguna untuk menggambarkan bentuk dari koordinatnya
#'figsize()' berguna untuk menentukan ukuran gambar

plt.plot(x,y, label = 'Grafik length' ,  color = 'purple' , 
         linewidth = 2 , marker = 'o' , markersize = 10 , 
         markeredgecolor = 'red' , linestyle='-')
#label untuk nama linier yang diminta
#linewidth untuk ketebalan linier
#marker untuk mejadi tanda pada titik yang dituju
#makersize unutuk ukuran tanda
#markeredgecolor untuk warna tanda

plt.plot(x1,y1, label = 'Garik width', marker = 'o' , 
         markersize = 10 , markeredgecolor = 'red')
#label untuk nama linier yang diminta
#marker untuk mejadi tanda pada titik yang dituju
#makersize unutuk ukuran tanda
#markeredgecolor untuk warna tanda

plt.title('Grafik Bunga', fontdict = {'fontname' : 'Georgia' , 'fontsize' : 15})
#title() untuk memasukan data dan ukuran font
#fontname untuk nama font yang diinginkan 
#fontdict untuk menentukan ukuran font

plt.xlabel('X', fontdict={'fontname' : 'Georgia' , 'fontsize' : 10})
plt.ylabel('Y', fontdict={'fontname' : 'Georgia' , 'fontsize' : 10})

plt.xticks(bunga['sepal length (cm)'])
plt.yticks(bunga['petal length (cm)'])
plt.legend()
#membentuk output dari label x dan y 

plt.savefig('grafik.png' , dpi=200)
#penyimpanan figure pada 'grafik.png' dengan resolusi 200 dpi

plt.show()
##show berfungsi untuk menampilkan grafik
```

# Histogram sederhana dengan Matplotlib

```python
bunga = pd.read_csv('https://raw.githubusercontent.com/imdwipayana/DATA/main/Bunga%20Iris/bunga_iris.csv')

petal = bunga['petal length (cm)']
petal1 = bunga['petal width (cm)']
#menginput data sesuai dengan kolomnya 
#menambah operator perbandingan sebagai pengelompokan data kolom yang diminta

plt.xlabel('Petal lenght (cm)')
plt.ylabel('Petal width (cm)')


print(plt.bar(petal,petal1))
#fungsi 'bar()' untuk membentuk histogram
```

# Properti Pembuatan Histogram dengan Matplotlib

```python
x = [-10,-5,0,5,10]
y = [20,10,0,10,20]
#menginput data integer 

plt.figure(figsize=(5,5))
#'figure()' berguna untuk menggambarkan bentuk dari koordinatnya
#'figsize()' berguna untuk menentukan ukuran gambar

hist = plt.bar(x,y,color='purple')
pola = ['+','+','+','+','+','+']
#menginisialisasikan variabel dengan fungsi 'bar' dan menentukan warna 

for bar in hist :
    bar.set_hatch(pola.pop(0))
#'for' untuk mengurutkann variabel hist
#'set_hatch()' berguna untuk pola di dalam batang
    
plt.show()
#show berfungsi untuk menampilkan histogram
```
